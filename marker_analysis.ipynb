```python
# Assign clusters to genes

assignation_dict = {
    'Zfhx3': ['1', '11'],
    'Zfhx3+Pou6f2': ['0','3','8'],
    'Foxp2': ['12', '7', '4', '5'],
    'Calb1': ['13'],
    'Nfib': ['2', '9']
}

reverse_assignation_dict = {}

for l in V1.obs['leiden'].cat.categories:
    reverse_assignation_dict[l] = 'other'
for g in assignation_dict.keys():
    for leid in assignation_dict[g]:
        reverse_assignation_dict[leid] = g

reverse_assignation_dict
```

```python
V1.obs['gene_assignation'] = V1.obs['leiden'].apply(lambda x:  reverse_assignation_dict[x])
```

```python
sc.pl.umap(V1, color='gene_assignation')
```
